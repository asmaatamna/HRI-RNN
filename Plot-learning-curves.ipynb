```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
```

```python
epochs = 100
tau = 5
eta = 2
hidden_dim = [32, 32]
lr = 1e-3
weight_decay = 1e-3
fold = 3
file_name = "HriRNN"
fname = "_" # "_robot_speaking_only_"
save = True
```

```python
# Load users data
train_losses = np.loadtxt("Train-losses_" + file_name + "_tau_" +
                          str(tau) + "_eta_" + str(eta) + "_" + str(epochs) + "_epochs_hdim_" +
                          str(hidden_dim) + "_lr_" + str(lr) + "_weight_decay_" + str(weight_decay) +
                          fname + str(fold) + ".txt")
validation_losses = np.loadtxt("Validation-losses_" + file_name + "_tau_" +
                         str(tau) + "_eta_" + str(eta) + "_" + str(epochs) + "_epochs_hdim_" +
                         str(hidden_dim) + "_lr_" + str(lr) + "_weight_decay_" + str(weight_decay) +
                         fname + str(fold) + ".txt")
```

```python
plt.suptitle("{} on Fold {}. Epochs: {}, hidden dim.: {}".format(file_name, fold, epochs, hidden_dim))

# First axis
ax1 = plt.subplot(211) # nrows, ncols, plot_number
ax1.plot(np.asarray(train_losses), label="Train")
ax1.set_ylabel("Training loss")
ax1.grid(True, which ='both')
ax1.set_ylim(ymin=0, ymax=0.125)
# ax1.legend(loc='best', fancybox=True, framealpha=0.5)

ax2 = plt.subplot(212) # nrows, ncols, plot_number
ax2.plot(np.asarray(validation_losses), label="Validation", color='orange')
ax2.set_ylabel("Validation loss")
ax2.grid(True, which ='both')
ax2.set_ylim(ymin=0.3, ymax=0.8)

plt.xlabel("Epochs")
plt.tight_layout(pad=2.5)
if save:
    plt.savefig("Learning_curves_" + file_name + "_tau_" + str(tau) + "_eta_" + str(eta) +
                "_" + str(epochs) + "_epochs_hdim_" + str(hidden_dim) + "_wd_" + str(weight_decay) +
                fname + str(fold) + ".pdf")
```
